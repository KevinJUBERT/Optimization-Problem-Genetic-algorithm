```python
def fonctionZ():
    
    check = False 
    
    while(check != True):
        
        Z = np.zeros(n)
        for i in range(n[0]):
            for j in range(n[1]):
                Z[i,j] = random.randint(0,1)
        
        L  = []
        for i in range(n[1]):
            dimension = np.sum(Z[:,i])
            if(dimension != 0):
                L.append(True)
        
        if(L == [True,True,True,True,True,True,True]):
            check = True

    return Z

#print(fonctionZ())
```

```python
def Z_func(yearly_demand, matrix_Z):

#    Z = np.array([1, 1, 0, 1])
    dimension = np.sum(matrix_Z)
    #print(dimension)
    
    if(dimension == 1):
        
        index_Z = list(matrix_Z).index(max(matrix_Z))
        matrix_Z[index_Z] = yearly_demand
    
    elif(dimension == 2):
        
        #Recupere l'indice des elements de la liste égale à 1
        index_list = []
        for i in range(len(matrix_Z)):
            if(matrix_Z[i] == 1):
                index_list.append(i)
        
        #Re-Organise randomly a list
        index_list = random.sample(index_list,2)

        #Affectation
        nombre1 = random.randint(0,int(yearly_demand))
        nombre2 = abs(yearly_demand - nombre1)
        
        #Index affectation
        index1 = index_list[0]
        index2 = index_list[1]
        
        #Affectation in matrix_Z
        matrix_Z[index1] = nombre1
        matrix_Z[index2] = nombre2
        
        
        
    elif(dimension == 3):
        
        #Recupere l'indice des elements de la liste égale à 1
        index_list = []
        for i in range(len(matrix_Z)):
            if(matrix_Z[i] == 1):
                index_list.append(i)
        
        #Re-Organise randomly a list
        index_list = random.sample(index_list,len(index_list))
        
        
        
        #Affectation
        nombre1 = random.randint(0,int(yearly_demand))
        nombre2 = random.randint(0,int(yearly_demand-nombre1))
        nombre3 = abs(yearly_demand - (nombre1+nombre2))
        
        #Index affectation
        index1 = index_list[0]
        index2 = index_list[1]
        index3 = index_list[2]
        
        #Affectation in matrix_Z
        matrix_Z[index1] = nombre1
        matrix_Z[index2] = nombre2
        matrix_Z[index3] = nombre3
        
        
    elif(dimension == 4):

        #Recupere l'indice des elements de la liste égale à 1
        index_list = []
        
        for i in range(len(matrix_Z)):
            if(matrix_Z[i] == 1):
                index_list.append(i)
        
        #Re-Organise randomly a list
        index_list = random.sample(index_list,len(index_list))
        
        
        
        #Affectation
        nombre1 = random.randint(0,int(yearly_demand))
        nombre2 = random.randint(0,int(yearly_demand-nombre1))
        nombre3 = random.randint(0,int(yearly_demand-(nombre1 + nombre2)))
        nombre4 = abs(yearly_demand - (nombre1+nombre2+nombre3)   )
        
        #Index affectation
        index1 = index_list[0]
        index2 = index_list[1]
        index3 = index_list[2]
        index4 = index_list[3]
     
        
        #Affectation in matrix_Z
        matrix_Z[index1] = nombre1
        matrix_Z[index2] = nombre2
        matrix_Z[index3] = nombre3
        matrix_Z[index4] = nombre4
        

                                 
    return matrix_Z
    
#test = Z_func(100, np.array([1,1,1,1]))
#print(yearly_demand)
#for i in range(100):
#    print(Z_func(160, np.array([1,1,0,0])))
```

Rééquilibre les excès ou déficites

```python
#print(check_demand(Child1))
#print(check_capacity(Child1))
#print("=============================== ")
#print(Child1)
#print("=============================== ")

def crossOver_condition(matrix):

    for j in range(n[1]):
        #print(check_demand(Child1)[1][j], end = " ")
        if (check_demand(matrix)[1][j] < 0) :
                index_faible  = check_capacity(matrix)[1].index(min(check_capacity(matrix)[1]))
                matrix[index_faible,j] =  matrix[index_faible,j] - check_demand(matrix)[1][j] 

        elif (check_demand(matrix)[1][j] > 0) :

                index_fort  = list(matrix[:,j]).index(max(matrix[:,j]))
                matrix[index_fort,j] =  matrix[index_fort,j] - check_demand(matrix)[1][j] 

    return matrix

#Child2 = crossOver_condition(Child2)


#print(check_demand(Child1))
#print(check_capacity(Child1))
#Child2
```

```python
def decomposition(limit):
    
    nombre = random.randint(1,limit-2)
    nombre2 = limit - nombre

    return [nombre,nombre2]
```
